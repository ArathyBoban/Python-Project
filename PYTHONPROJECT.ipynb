```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('myexcel.csv')

print(data)
```

```python
#question 1
```

```python
team_distribution = data['Team'].value_counts()
team_percentage = (team_distribution / team_distribution.sum()) * 100

plt.figure(figsize=(8, 6))
team_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Employees Across Teams')
plt.ylabel('')
plt.show()
```

```python
#question 2
```

```python
position_distribution = data['Position'].value_counts()
plt.figure(figsize=(8, 6))
position_distribution.plot(kind='bar', color='skyblue')
plt.title('Number of Employees by position')
plt.xlabel('position')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
#question 3
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55','56-65']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

age_group_distribution = data['Age Group'].value_counts()

plt.figure(figsize=(8, 6))
age_group_distribution.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()
```

```python
#question 4
```

```python
team_salary = data.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = data.groupby('Position')['Salary'].sum().sort_values(ascending=False)

#Visualisation: Salary by Team
plt.figure(figsize=(8, 6))
team_salary.plot(kind='bar', color='orange')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.xticks(rotation=45)
plt.show()

# Visualisation: Salary by Position
plt.figure(figsize=(8, 6))
position_salary.plot(kind='bar', color='purple')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#question 5
```

```python
correlation = data['Age'].corr(data['Salary'])
print(f'Correlation between Age and Salary: {correlation}')

plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Age'], y=data['Salary'])
sns.regplot(x=data['Age'], y=data['Salary'],scatter=False, color='red')
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
